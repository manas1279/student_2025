# Homework: Ultimate Battle

---

## Scenario

You are programming a game where 2 objects of your choosing (e.g., Robots, Dinosaurs, People) battle. Each object has health and power.

---

## Instructions

1. **Create a class** representing your object with:

   **Instance variables:**  
   - `String name`  
   - `int power`  
   - `int health`  
   - `MAKE YOUR OWN`

   **Static variables:**
   - `double fightDuration`

   **Instance methods:**  
   - `void attack()`
   - `void printStatus()`
   - `MAKE YOUR OWN`

   **Class methods:**  
   - `static int strongerFighter()`
   - `static void beginBattle()`
   - `MAKE YOUR OWN`

2. **In `main`:**  
   - Create two objects.  
   - Use instance methods to attack and print status.  
   - Use static methods to compare, print a fact, and start the battle.

```java
public class UltimateBattle {

    // ‚öôÔ∏è Robot Class
    static class Robot {
        // Instance variables
        String name;
        int power;
        int health;
        String weapon;

        // Static variable
        static double fightDuration = 0;

        // Constructor
        public Robot(String name, int power, int health, String weapon) {
            this.name = name;
            this.power = power;
            this.health = health;
            this.weapon = weapon;
        }

        // Instance method: attack another robot
        public void attack(Robot enemy) {
            System.out.println(this.name + " attacks " + enemy.name + " with " + weapon + "!");
            enemy.health -= this.power;
            if (enemy.health < 0) enemy.health = 0;
            fightDuration += 0.5;
        }

        // Instance method: print status
        public void printStatus() {
            System.out.println(name + " | Health: " + health + " | Power: " + power + " | Weapon: " + weapon);
        }

        // Extra instance method: check if robot is still alive
        public boolean isAlive() {
            return health > 0;
        }

        // Static method: compare who is stronger
        public static Robot strongerFighter(Robot r1, Robot r2) {
            if (r1.power > r2.power) return r1;
            else if (r2.power > r1.power) return r2;
            else return null;
        }

        // Static method: begin the battle
        public static void beginBattle(Robot r1, Robot r2) {
            System.out.println("\n‚öîÔ∏è Battle begins between " + r1.name + " and " + r2.name + "!");
            fightDuration = 0;

            // Battle loop
            while (r1.isAlive() && r2.isAlive()) {
                r1.attack(r2);
                if (!r2.isAlive()) break;
                r2.attack(r1);
            }

            // End of battle
            System.out.println("\nüèÅ Battle ended in " + fightDuration + " rounds.");
            if (r1.isAlive()) System.out.println("üí™ " + r1.name + " wins!");
            else System.out.println("üí• " + r2.name + " wins!");
        }

        // Static fun fact
        public static void printFact() {
            System.out.println(" Fun Fact: Robots never need sleep, only maintenance!");
        }
    }

    //  Main method
    public static void main(String[] args) {
        // Create two robots
        Robot alpha = new Robot("Alpha", 25, 100, "Laser Blaster");
        Robot beta = new Robot("Beta", 20, 110, "Rocket Punch");

        // Print their status
        alpha.printStatus();
        beta.printStatus();

        // Compare power
        Robot stronger = Robot.strongerFighter(alpha, beta);
        if (stronger != null)
            System.out.println("\n‚ö° Stronger fighter: " + stronger.name);
        else
            System.out.println("\n‚ö° Both fighters are equally strong!");

        // Fun fact
        Robot.printFact();

        // Start the battle
        Robot.beginBattle(alpha, beta);

        // Print final status
        System.out.println("\nFinal Status:");
        alpha.printStatus();
        beta.printStatus();
    }
}
```
